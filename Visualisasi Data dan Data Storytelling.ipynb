## Visualisasi Data

Visualisasi data adalah salah satu cara utama di mana kita dapat memeriksa data dan mendapatkan insight darinya.

Mari kita coba beberapa jenis grafik yang sering digunakan untuk visualisasi data.

**Diagram Batang**
Diagram batang adalah cara yang baik untuk membandingkan jumlah atau jumlah numerik di seluruh kategori. 

```python
#Hitung jumlah responden berdasarkan preferensi pemilihan jenis makanan berdasarkan faktor agama
jumlah_resp_faktor_agama = data['Religious Factor of Food Type'].value_counts()

#Membuat diagram batang
jumlah_resp_faktor_agama.plot(kind='bar', title='Opini Responden terhadap Faktor Agama dalam Pemilihan Makanan')
```

Pada diagram batang di atas, maka Anda akan melihat bahwa mayoritas responden cenderung mempertimbangkan faktor-faktor religi dalam pemilihan makanan (mungkin halal/haram).


**Diagram Garis**

Diagram garis adalah diagram yang bagus untuk melihat perubahan nilai di sepanjang data, biasanya (tetapi tidak selalu) berdasarkan periode waktu. Pada dataset perilaku konsumsi di area urban di Indonesia tidak tersedia data jenis ini, sehingga kita menggunakan data kasus harian covid-19 di Indonesia yang diperoleh dari https://github.com/owid/covid-19-data. Berikut contoh code dan hasil visualisasi dari data harian kasus baru Covid-19 di Indonesia sejak maret 2020 hingga Oktober 2021.

```python
#Load data (Data dengan ekstensi file txt jika pemisah datanya berupa koma "," maka dapat dibaca dengan read_csv)
df = pd.read_csv('data/new_cases.txt')

#Membuat grafik garis untuk kolom Indonesia
df['Indonesia'].plot(kind='line', title='Perkembangan Jumlah Kasus Harian Covid-19 Indonesia',figsize=(10,4))
```

Jika Anda sudah mempelajari mengenai machine learing, Anda memerlukan proses EDA (Exploratory Data Analysis) yang salah satunya adalah melihat apakah ada korelasi yang terbentuk dari 2 fitur yang kita miliki sebelum kita membuat model machine learning. Apa hal menarik yang dapat disimpulkan dari diagram di atas?

**Histogram**

Diagram batang berfungsi dengan baik untuk membandingkan nilai numerik kategorikal atau diskrit. Saat Anda perlu membandingkan nilai kuantitatif kontinu, Anda dapat menggunakan bentuk bagan serupa yang disebut histogram. Histogram berbeda dari bagan batang dalam hal mereka mengelompokkan nilai berkelanjutan ke dalam rentang, jadi bagan tidak menunjukkan batang untuk setiap nilai individual, melainkan batang untuk setiap rentang nilai yang dikumpulkan. Karena bagan ini mewakili data berkelanjutan daripada data diskrit, batang tidak dipisahkan oleh celah. Biasanya, histogram digunakan untuk menunjukkan frekuensi relatif nilai dalam kumpulan data.

Berikut contoh implementasi pembuatan Histogram di Python untuk kolom **Years Lived in Urban Area**:

```python
data['Years Lived in Urban Area'].plot(kind='hist',bins=10, title='Jumlah Responden Berdasarkan Lama Tinggal di Kota (Tahun)')
```

Apa insight yang dapat diambil dari diagram di atas?

**Bagaimana membangun visualisasi dari data dan merangkai menjadi sebuah cerita?**

Seringkali kita dibuat bingung mengenai apa saja yang bisa divisualisasikan dari sebuah data yang mungkin kolomnya banyak. Seperti yang dijelaskan pada slide, bahwa kita dapat beranjak dari penentuan topik umum yang ingin dibahas atau dijawab permasalahannya lalu jabarkan ke beberapa pertanyaan yang jawabannya dapat dijawab dengan visualisasi data. Contoh sebagai berikut:


Topik: Preferensi dan Perilaku Konsumsi Makanan di area Urban di Indonesia

Pertanyaan:
1. Kota mana dengan rata-rata % pengeluaran makan paling besar?
2. Bagaimana perilaku pemilihan makanan berdasarkan harga terhadap social class masyarakat?
3. Apakah tingkat pendidikan sarjana memiliki preferensi memilih makanan-makanan yang sehat?
4. Apakah warga DKI Jakarta masih mengonsumsi makanan tradisional?
5. Usia berapa saja yang masih mengonsumsi makanan tradisional?

```python
#Kota mana dengan rata-rata % pengeluaran makanan paling besar?

fig1,ax1 = plt.subplots() #Definisikan figure dan sumbu (axes), analogi figure adalah kertas, ax adalah sumbu tempat kita menggambar grafik

data.groupby('City').mean()['% Monthly Food Expense'].sort_values(ascending=False).plot(kind='bar',ax=ax1)

ax1.set_title('Rata-Rata % Pengeluaran Makanan Tiap Kota') #set judul grafik
ax1.set_xlabel('Kota') #Set keterangan sumbu x
ax1.set_ylabel('% Pengeluaran Makanan terhadap Total Pengeluaran') #Set keterangan sumbu y
```

Bagaimana interpretasi dari grafik di atas?

```python
#Bagaimana perilaku pemilihan makanan berdasarkan faktor harga terhadap social class masyarakat?

data['Social Class'].unique()
```

```python
fig2,ax2 = plt.subplots(ncols=3,nrows=1,figsize=(24,5)) #ncols adalah jumlah kolom dan nrows adalah jumlah baris. Output ax2 adalah list yang jumlah elemennya sebanyak ncols*nrows

lower_class = data[data['Social Class']=='Lower Class'] #Seleksi data yang diperlukan
group_low = lower_class['Price Factor'].value_counts() #Hitung frekuensi respon price factor
group_low.plot(kind='pie',ax=ax2[0])

middle_class = data[data['Social Class']=='Middle Class']
group_mid = middle_class['Price Factor'].value_counts()
group_mid.plot(kind='pie',ax=ax2[1])

upper_class = data[data['Social Class']=='Upper Class']
group_up = upper_class['Price Factor'].value_counts()
group_up.plot(kind='pie',ax=ax2[2])

ax2[0].set_title('Price Food dan Lower Class')
ax2[1].set_title('Price Food dan Middle Class')
ax2[2].set_title('Price Food dan Upper Class')
```

Apa yang dapat Anda ceritakan dari ketiga diagram di atas?

```python
# Apakah warga DKI Jakarta masih mengonsumsi makanan tradisional?

data[data['Province']=='DKI Jakarta']['Traditional Food Practicing'].value_counts().plot.barh(title='Preferensi Makanan Tradisional Warga DKI Jakarta', color='red') #Diagram batang horizontal
```

Apa jawaban yang dapat menjawab pertanyaan "Apakah warga DKI jakarta masih mengonsumsi makanan tradisional?"

```python
# Usia berapa saja yang masih mengonsumsi makanan tradisional?

data[(data['Traditional Food Practicing']=='Strongly Agree') | (data['Traditional Food Practicing']=='Agree')]['Age'].plot(kind='hist',color='green')
```
